```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# Carregar o DataFrame
df = pd.read_csv('datasets/imdb-reviews-pt-br.csv')

# Verificar o nome das colunas
print(df.columns)

def calcular_polaridade(comentario):
    return TextBlob(comentario).sentiment.polarity

# Substituir 'comments' pelo nome correto da coluna de resenhas
df['polaridade'] = df['text_pt'].apply(calcular_polaridade)

# Exibir os comentários e suas polaridades
print(df[['text_pt', 'polaridade']])
```

```python
from sklearn.model_selection import train_test_split
```

```python
treino, teste, classe_treino, classe_teste = train_test_split(
    df_resenhas.text_pt, df_resenhas.sentiment, random_state=42
)
```

# Escolhendo algoritmo de aprendizado supervisionado -
Classificação

```python
from sklearn.linear_model import LogisticRegression

# Criando um objeto Chamada do construtor da classe LogisticRegression
regressao_logistica = LogisticRegression()

# treinamento
regressao_logistica.fit(treino, classe_treino) # rótulos - neg pos

# acurácia
acuracia = regressao_logistica.score(teste, classe_teste)
print(f"Acurácia: {acuracia}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer


tfidf = TfidfVectorizer()
treino_numerical = tfidf.fit_transform(treino)

# Criando um objeto Chamada do construtor da classe LogisticRegression
regressao_logistica = LogisticRegression()

# treinamento
regressao_logistica.fit(treino_numerical, classe_treino) # rótulos - neg pos

# Pré-processamento
teste_numerical = tfidf.transform(teste)

# acurácia
acuracia = regressao_logistica.score(teste_numerical, classe_teste)
print(f"Acurácia: {acuracia}")
```
