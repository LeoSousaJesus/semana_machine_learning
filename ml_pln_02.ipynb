# Escolhendo algoritmo de aprendizado supervisionado -

```python
from sklearn.linear_model import LogisticRegression

# Criando um objeto Chamada do construtor da classe LogisticRegression
regressao_logistica = LogisticRegression()

# treinamento
regressao_logistica.fit(treino, classe_treino) # rótulos - neg pos

# acurácia
acuracia = regressao_logistica.score(teste, classe_teste)
print(f"Acurácia: {acuracia}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer


tfidf = TfidfVectorizer()
treino_numerical = tfidf.fit_transform(treino)

# Criando um objeto Chamada do construtor da classe LogisticRegression
regressao_logistica = LogisticRegression()

# treinamento
regressao_logistica.fit(treino_numerical, classe_treino) # rótulos - neg pos

# Pré-processamento
teste_numerical = tfidf.transform(teste)

# acurácia
acuracia = regressao_logistica.score(teste_numerical, classe_teste)
print(f"Acurácia: {acuracia}")
```

# Bag of Words - Prática 01

image.png

```python
# vetorização - sklearn (CountVectorizer)

#CountVectorizer -  Vetorizar o Corpus em Números
from sklearn.feature_extraction.text import CountVectorizer

# Corpus
texto_bag = ["o filme é muito bom", "o filme é muito ruim"]

# Vetorização
vetorizar = CountVectorizer(lowercase=False)

# Criando nossa Bag of Words (BoW)
bag_of_words = vetorizar.fit_transform(texto_bag)
```

```python
bag_of_words
```

```python
vetorizar.get_feature_names_out()
```

```python
vetor = vetorizar.fit_transform(texto_bag)
vetor = vetor.todense()
print(vetor)
```

```python
matriz = pd.DataFrame(bag_of_words.toarray(), columns=vetorizar.get_feature_names_out())
matriz
```

```python
matriz_esparsa = pd.DataFrame.sparse.from_spmatrix(bag_of_words, columns=vetorizar.get_feature_names_out())
matriz_esparsa
```

toarray() - conversão considerando uma matriz densa (inclui tudo)

sparse.from_matrix()- cria DataFrame sparso diretamente da matriz sparsa.

Vantagem:


1.   Eficiência
2.   Memória
3.   Especializa em  muitos zeros
4.   Performance

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Corpus
texto_bag = ["o filme é muito bom", "o filme é muito ruim"]

# Vetorização
vetorizar = CountVectorizer(lowercase=False)

# Criando nossa Bag of Words (BoW)
bag_of_words = vetorizar.fit_transform(texto_bag)

# Convert the sparse matrix to a dense array
bag_of_words_dense = bag_of_words.toarray()

# Create the DataFrame using the dense array
matriz = pd.DataFrame(bag_of_words_dense, columns=vetorizar.get_feature_names_out())

matriz
```
