## Essa tabela nos ajudará a entender o problema de aprendizado de máquina e classificar se um determinado animal é uma galinha ou um papagaio 

# Lista com papagaios

```python
# Papagaio

# features [1 sim, 0 não]
# tem bico?
# risco de extinção?
# faz cocoricó

papagaio1 = [1,1,0]
papagaio2 = [1,1,0]
papagaio3 = [1,1,1]
```

# Lista com as galinhas

```python
# Galinha

# features [1 sim, 0 não]
# tem bico?
# risco de extinção?
# faz cocoricó

galinha1 = [1,0,1]
galinha2 = [1,0,1]
galinha3 = [1,1,1]
```

```python
dados = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
```

```python
dados = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
classes = [1,1,1,0,0,0]
```

```python
dados
```

```python
! pip install scikit-learn
```

```python
from sklearn.svm import LinearSVC

modelo = LinearSVC()
```

```python
modelo.fit(dados, classes)
```

# Prever um animal

```python
# features [1 sim, 0 não]
# tem bico?
# risco de extinção?
# faz cocoricó

# tem bico, está em extinção e não faz cocoricó

animal_misterioso = [1,1,0]
previsao = modelo.predict([animal_misterioso])
```

```python
previsao
```

```python
# features [1 sim, 0 não]
# tem bico?
# risco de extinção?
# faz cocoricó

# tem bico, está em extinção e não faz cocoricó

animal_misterioso2 = [1,0,1]
previsao2 = modelo.predict([animal_misterioso2])
```

# Calculando Acurácia do Modelo
## Dados e Rótulos

```python
# dados do treino x
treino_x = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
#  dados do treino y
treino_y = [1,1,1,0,0,0]
```

```python
# animais misteriosos - para teste
misterioso1 = [1,1,1]
misterioso2 = [1,1,0]
misterioso3 = [0,1,1]
```

```python
# previsões - dados de testes
teste_x = [misterioso1, misterioso2, misterioso3]
# prevendo  os testes
previsoes = modelo.predict(teste_x)
```

```python
# rótulos (classes de teste) (avaliação - aqui é a sua provinha)
teste_y = [0,1,1]
```

```python
# importando módulo que já traz a função de acurácia encapsulada
from sklearn.metrics import accuracy_score

# definindo uma variável para receber o retorno da função acurácia * 100 para porcentagem
taxa_de_acerto = accuracy_score(teste_y, previsoes) 

# mostrando a acúracia
print(f"Acurácia: {taxa_de_acerto * 100:.2f}%")
```
