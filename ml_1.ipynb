## Essa tabela nos ajudar a entender o problema de aprendizado de mquina e classificar se um determinado animal  uma galinha ou um papagaio 

# Lista com papagaios

```python
# Papagaio

# features [1 sim, 0 no]
# tem bico?
# risco de extino?
# faz cocoric

papagaio1 = [1,1,0]
papagaio2 = [1,1,0]
papagaio3 = [1,1,1]
```

# Lista com as galinhas

```python
# Galinha

# features [1 sim, 0 no]
# tem bico?
# risco de extino?
# faz cocoric

galinha1 = [1,0,1]
galinha2 = [1,0,1]
galinha3 = [1,1,1]
```

```python
dados = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
```

```python
dados = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
classes = [1,1,1,0,0,0]
```

```python
dados
```

```python
! pip install scikit-learn
```

```python
from sklearn.svm import LinearSVC

modelo = LinearSVC()
```

```python
modelo.fit(dados, classes)
```

# Prever um animal

```python
# features [1 sim, 0 no]
# tem bico?
# risco de extino?
# faz cocoric

# tem bico, est em extino e no faz cocoric

animal_misterioso = [1,1,0]
previsao = modelo.predict([animal_misterioso])
```

```python
# features [1 sim, 0 no]
# tem bico?
# risco de extino?
# faz cocoric

# tem bico, est em extino e no faz cocoric

animal_misterioso2 = [1,0,1]
previsao2 = modelo.predict([animal_misterioso2])
```

# Calculando Acurcia do Modelo
## Dados e Rtulos

```python
# dados do treino x
treino_x = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
#  dados do treino y
treino_y = [1,1,1,0,0,0]
```

```python
# animais misteriosos - para teste
misterioso1 = [1,1,1]
misterioso2 = [1,1,0]
misterioso3 = [0,1,1]
```

```python
# previses - dados de testes
teste_x = [misterioso1, misterioso2, misterioso3]
# prevendo  os testes
previsoes = modelo.predict(teste_x)
```

```python
# rtulos (classes de teste) (avaliao - aqui  a sua provinha)
teste_y = [0,1,1]
```

```python
# importando mdulo que j traz a funo de acurcia encapsulada
from sklearn.metrics import accuracy_score

# definindo uma varivel para receber o retorno da funo acurcia * 100 para porcentagem
taxa_de_acerto = accuracy_score(teste_y, previsoes) 

# mostrando a acracia
print(f"Acurcia: {taxa_de_acerto * 100:.2f}%")
```
